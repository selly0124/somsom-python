x / (y+z)     +      y / (x+z)     + z / (x+y) = 4

x(x+z)(x+y)	+ y(x+y)(y+z) + z(x+z)(y+z)

```python
x = 1
y = 5
z = 7
```

```python
sum = (x+y) * (y+z) * (x+z)
result = x * (x+z) * (x+y) + y * (x+y) * (y+z) + z * (x+z) * (y+z)
answer = result / sum
```

```python
print(answer)
```

```python
def calcresult(x,y,z):
    sum = (x+y) * (y+z) * (x+z)
    result = x * (x+z) * (x+y) + y * (x+y) * (y+z) + z * (x+z) * (y+z)
    
    if sum != 0:
        res = result / sum
    else:
        res = 0
    return (res)
```

```python
for i1 in range(-100,100,1):
    for i2 in range (1000):
        for i3 in range(1000):
            x = i1 + 1
            y = i2 + 1
            z = i3 + 1
            
            res = calcresult(x,y,z)
            if y == 1000 & z == 1000:
                print(f'epoch {x} running...')
            if res > 3.99999:
                if res < 4.00001:
                    print(f'근사해 >>>>>>>>>>>>>>>> x : {x}   y : {y}    z : {z}   result : {res}')
            if res ==  4:
                print(f'정답 >>>>>>>>>>>>>>>>> x : {x}   y : {y}    z : {z}   result : {res}')

```
